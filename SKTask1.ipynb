```python
# Heart rate Variability for individual
import matplotlib.pyplot as plt2
def plot_hrv(row):
    heartrate_100 = row.heartrate_100
    time_100 = row.time_100
    l=[]
    t=[]
    l=heartrate_100
    t=time_100
    plt2.figure(figsize=(30,6))
    plt2.plot(t,l,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=5)
    plt2.title('time vs Heartrate')
    plt2.xlabel('Time')
    plt2.ylabel('BPM')
    
plot_hrv(df.iloc[0])
```

```python
#detecting peaks
def detect_peaks(heartrate, threshold=0.3, qrs_filter=None):
    
    if qrs_filter is None:
        # create default qrs filter, which is just a part of the sine function
        t = np.linspace(1.5 * np.pi, 3.5 * np.pi, 15)
        qrs_filter = np.sin(t)
    heartrate = np.array(heartrate)
    
    # normalize data
    heartrate = (heartrate - heartrate.mean()) / heartrate.std()

  
    similarity = np.correlate(heartrate, qrs_filter, mode="same")
    similarity = similarity / np.max(similarity)
    
    heartrate =   heartrate.tolist()
    peaks=[]
    for i in range(len(heartrate)):
        if heartrate[i]>threshold:
            peaks.append(i)
    return peaks, similarity

def plot_hrp(row):
    heartrate_100 = row.heartrate_100
    time_100 = row.time_100
    l=[]
    t=[]
    l=heartrate_100
    t=time_100                     
    peaks, similarity = detect_peaks(l, threshold=0.3)
    # Extract Peak points in heart rate
    # plot 3 way
                    
    a=[]
    c=[]
    for i in range(len(peaks)):
        b=peaks[i]
        a.append(l[b])
    for i in range(len(peaks)):
        b=peaks[i]
        c.append(t[b])
    plt.figure(figsize=(20, 15))
    
    plt.subplot(211)
    plt.title("Heart rates with  peaks")
    plt.plot(c, a, label="Heartrate", color="#51A6D8",marker="o", linewidth=1)
    plt.legend(loc="upper right")
    plt.xlabel("Time ")
    plt.ylabel("Heartrate")  
    plt.subplot(212)
    plt.title('Similarity with QRS ')
    plt.plot(t, similarity, label="Similarity with QRS ", color="olive", linewidth=1)
    plt.legend(loc="upper right")
    plt.xlabel("Time ")
    plt.ylabel("Similarity (normalized)")
    

    
plot_hrp(df.iloc[0])
```

```python
# Plotting Heart rate and indicating min max sd avg
import statistics
def plot_hrm(row):
    heartrate_100 = row.heartrate_100
    time_100 = row.time_100
    #plotting maximun
    ymax = max(heartrate_100)
    xmax = time_100[np.argmax(heartrate_100)]
    
    #plotting minimum
    ymin = min(heartrate_100)
    xmin = time_100[np.argmin(heartrate_100)]
    
    mu=mean(heartrate_100)
    sigma= statistics.stdev(heartrate_100)
    
    
    
    # SD zone
    
    
    fig, ax1 = plt.subplots(figsize=(18,8))
    ax1.plot(time_100, heartrate_100, color='r')
    upper = [mu+sigma]*429
    lower= [mu-sigma]*429
    avg=[mu]*429
    
    ax1.plot(time_100, upper, color='black',linestyle='dashed')
    ax1.plot(time_100, lower, color='blue',linestyle='dashed')
    ax1.plot(time_100, avg, color='green')
    
    ax1.annotate('Maximum Heartrate', xy=(xmax, ymax), xytext=(xmax, ymax+5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
    ax1.annotate('Minimum Heartrate', xy=(xmin, ymin), xytext=(xmin, ymin+5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
    #ax1.fill_between(time_100, mu+sigma, mu-sigma,facecolor='blue', alpha=0.5)

    return

plot_hrm(df.iloc[6])
```
